```python
import random
```

```python
def get_user_choice():
  user_choice  = input("select one rock , paper or scissor ")
  print( user_choice )
  return user_choice
```

```python
def get_computer_choice():
  options = ["rock" , "paper" , "scissor"]
  computer_choice = random.choice(options)
  print(computer_choice)
  return computer_choice
```

```python
def select_winner():
  uc = get_user_choice()
  cc = get_computer_choice()

  if uc == cc:
    print("It's a tie!")
  elif (uc == "rock" and cc == "scissor") or \
       (uc == "paper" and cc == "rock") or \
       (uc == "scissor" and cc == "paper"):
    print("You win!")
  else:
    print("Computer wins!")
```

```python
select_winner()
```
